## MISSING VALUE TREATMENT - MEAN

```python
marks=[91,96,87,76,56,45,34,56,78,90]
np.mean(marks)
```

```python
marks=[91,96,87,76,56,45,34,np.nan,56,78,90,np.nan,np.nan,np.nan]
marks
```

```python
import pandas as pd
marks=[91,96,87,76,56,45,34,np.nan,56,78,90,np.nan,np.nan,np.nan]
marks_series=pd.Series(marks)
marks_series.fillna(marks_series.mean())
```

```python
marks=[91,96,87,76,56,45,34,56,78,90,1000000]
np.mean(marks)
```

```python
marks=[91,96,87,76,56,45,34,np.nan,56,78,90,1000000,np.nan,np.nan,np.nan]
marks
```

```python
import pandas as pd
marks=[91,96,87,76,56,45,34,np.nan,56,78,90,100000,np.nan,np.nan,np.nan]
marks_series=pd.Series(marks)
marks_series.fillna(marks_series.median())
```

```python
marks=['red','green','blue','red',np.nan,'green','red',np.nan]
marks_series=pd.Series(marks)
marks_series.fillna(marks_series.mode()[0])
```
